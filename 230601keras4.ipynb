```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.utils import to_categorical
```

```python
(X_train, y_train),(X_test, y_test) = mnist.load_data()
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
model = Sequential(
    [
        Flatten(input_shape=(28,28)),
        Dense(256, activation='relu'),
        Dense(128, activation='relu'),
        Dense(10, activation='softmax'),
    ]
)
```

```python
model.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train,
         y_train,
         epochs=20,
         batch_size=64,
         validation_data = (X_test, y_test),
         shuffle = True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
val_loss = model.history.history['val_loss']
val_acc = model.history.history['val_accuracy']
plt.plot(np.arange(1,21), val_loss)
```

```python
earlystop = EarlyStopping(patience=10, monitor='val_loss')
checkpoint = ModelCheckpoint(filepath='./best_model/', save_best_only=True)
tensorboard = TensorBoard(log_dir='./log-dir')
```

```python
model.fit(X_train,
         y_train,
         epochs=20,
         batch_size=64,
         validation_data = (X_test, y_test),
         shuffle = True,
         callbacks = [earlystop, checkpoint, tensorboard]
         )
```
