```python
x = np.random.randint(1,11, size=(100,5))
x = x.astype('object')
```

```python
for _ in range(20):
    row = np.random.randint(0,100,1)
    col = np.random.randint(0,5,1)
    x[row, col] = np.nan
# x
```

```python
x = pd.DataFrame(x)
```

```python
x[x.isna().sum(axis=1) > 0]
```

```python
x.fillna(method='ffill')
```

```python
x = np.random.randint(50, 100, size=(100,4))
x = pd.DataFrame(x)
x.columns = ['col_'+ str(i+1) for i in range(4) ]
x['gender'] = np.random.choice(np.array(['f', 'm']), 100)
```

```python
x.groupby('gender').mean()[['col_1','col_2']]
```

```python
x.groupby('gender').mean()[['col_1','col_2']].unstack()
```
