```python
def cross_val_sc(model, X, y, cv=5):
    fold =KFold(cv)
    score = []
    for train_ind, test_ind in fold.split(X):
        model = model
        X_train = X[train_ind]
        X_test = X[test_ind]
        y_train = y[train_ind]
        y_test = y[test_ind]
        model.fit(X_train, y_train)
        score.append(model.score(X_test, y_test))
    score = np.array(score)
    score_mean = np.mean(score)
    return score, score_mean
```

```python
# grid = GridSearchCV(model, params, cv=5)
import itertools
def cross_val_sc(model, X, y, cv=5):
    fold =KFold(cv)
    score = 0
    for train_ind, test_ind in fold.split(X):
        model = model
        X_train, y_train = X[train_ind], y[train_ind]
        X_test, y_test = X[test_ind], y[test_ind]
        model.fit(X_train, y_train)
        score += model.score(X_test, y_test)   
    score_mean = np.around(score/cv,3)
    return score, score_mean

class GridSearchParam:
    def __init__(self,model,params,cv=5):
        self.model = model
        self.params = params
        self.cv = cv

    def fit(self,x,y):
        self.score = []
        self.combinations = list(itertools.product(*params.values()))
        self.param_combinations = [{k: v for k, v in zip(params.keys(), combination)}\
                              for combination in self.combinations]

        self.score = []
        for v in param_combinations:
            model = self.model(**v)
            cross_val_sc(model, x, y)
            self.score.append((v, cross_val_sc(model, x, y)))
        self.best_estimator_ = self.score[np.argmax(np.array([v for i, v in grid.score]))][0]
        self.best_model = self.model(**self.best_estimator_)
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```
