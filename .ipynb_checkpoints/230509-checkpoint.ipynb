```python
from functools import cache
```

```python
@cache
def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1) + fibo(n-2)
```

```python
def cost(W, b):
    y_hat = sigmoid(np.dot(X,W)+b)
    return -np.sum(y*np.log(y_hat) + (1-y)*np.log(1-y_hat))
```

```python
def diff(func,W,b):
    h = 1e-4
    grad_a = np.zeros_like(W)
    for i in range(W.shape[0]):
        for j in range(W.shape[1]):
            tmp = W[i,j]
            fx = func(W, b)
            W[i,j] = tmp + h
            fxh = func(W,b)
            grad_a[i,j] = (fxh - fx)/h
            W[i,j] = tmp
    fx = func(W,b)
    fxh = func(W, b+h)
    grad_b = (fxh - fx)/h

    return grad_a, grad_b
```

```python
lr = 1e-3
epochs = 10000
for i in range(epochs):
        W -= diff(cost,W,b)[0]*lr
        b -= diff(cost,W,b)[1]*lr
```

```python
def camelcase(x):
    return ''.join([ word.capitalize() for word in x.split('_')])
```

```python
camelcase('asia_seoul')
```

```python
# camelcase(['asia_seoul', 'asia_tokyo', 'europe_paris']) ## error 발생
```

```python
def mapper(func):
    def wrapper(list_arg):
        return [ func(word) for word in list_arg]
    return wrapper

@mapper
def camelcase(x):
    return ''.join([ word.capitalize() for word in x.split('_')])
```

```python
x = ['asia_seoul', 'europe_paris', 'america_usa']
camelcase(x)
```

```python
## 스케일링 함수
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
def min_max(x):
    return (x - min(x))/(np.max(x) - np.min(x))
def standard(x):
    return (x - np.mean(x)) / np.sqrt(np.mean((x-np.mean(x))**2))
def robust(x):
    return (x - np.percentile(x, 50))/(np.percentile(x, 75) - np.percentile(x, 25))
```

```python
## one hot 만드는 함수
def make_onehot(x):
    col = np.unique(x).size
    row = len(x) ## x.shape[0]
    onehot = np.zeros((row, col))
    for idx, val in enumerate(x):
        onehot[idx, val] = 1
    return onehot

## softmax 함수작성
def softmax(x):
    m = np.max(x)
    x -= m
    return np.exp(x)/np.sum(np.exp(x))

def sigmoid(x):
     return 1/(1+np.exp(-x))
    
class Leejunyoung:
    def __init__(self,x,y):
        self.x = x
        self.y_original = y
        self.y = make_onehot(y)
        # self.w = np.random.randn(self.x.shape[1],self.y.shape[1])
        self.w = np.random.uniform(0,1,size=(self.x.shape[1], self.y.shape[1]))
        self.b = np.zeros(self.y.shape[1])
        
    def fit(self, epochs,lr):
        for epoch in range(epochs):
            self.w -= self.diff()[0]*lr
            self.b -= self.diff()[1]*lr
            if epoch % 10 == 0:
                print(f'Loss : =======> {self.cost()}')
    
    def predict(self):
        self.y_hat = np.dot(self.x, self.w) + self.b
        self.y_hat = sigmoid(self.y_hat)
        self.y_hat = np.apply_along_axis(softmax, 1, self.y_hat)
        return self.y_hat
        
    def cost(self):
        self.predict()
        epsilon = 1e-7
        return -np.sum(self.y*np.log(self.y_hat+epsilon))
    
    def diff(self):
        h = 1e-4
        self.grad_a = np.zeros_like(self.w)
        for i in range(self.w.shape[0]):
            for j in range(self.w.shape[1]):
                tmp = self.w[i,j]
                fx = self.cost()
                self.w[i,j] += h
                fxh = self.cost()
                self.grad_a[i,j] = (fxh - fx)/h
                self.w[i,j] = tmp
        self.grad_b = np.zeros_like(self.b)
        for i in range(self.b.shape[0]):
            tmp = self.b[i]
            fx = self.cost()
            self.b[i] += h
            fxh = self.cost()
            self.grad_b[i] = (fxh-fx)/h
            self.b[i] = tmp
        return self.grad_a, self.grad_b
    
    def score(self, x, y):
        predict = np.argmax(softmax(np.dot(x, self.w)+self.b), 1)
        score = np.sum(predict == y)/len(x)
        return score
```

```python
from sklearn.datasets import load_iris, load_wine
# X = load_iris()['data']
# y = load_iris()['target']
X = load_wine()['data']
y = load_wine()['target']
```

```python
X = np.apply_along_axis(robust, 1 ,X)
```
