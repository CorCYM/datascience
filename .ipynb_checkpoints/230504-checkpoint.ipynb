```python
class Calc:
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.add()
        
    def add(self):
        self.add_result = self.x + self.y
        return self.add_result
    
    def __call__(self):
        return self.x * self.y
    
```

```python
from sklearn.datasets import load_iris
X = load_iris().data
y = load_iris().target
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```

```python
# 수치미분
def diff(f,x):
    h = 1e-4
    fxh = f(x+h)
    fx = f(x)
    return (fxh - fx)/h
```

```python
def func(x):
    return 2*x**2 + 3*x + 4
```

```python
## 편의를 위해 x,y를 조절
x = np.random.randint(1,30,1000)
a = 257
y = x*a
```

```python
plt.plot(x,y)
```

```python
a_hat = np.random.randn(1)
```

```python
## 해석미분
for _ in range(500):
    # print(a_hat)
    y_hat = a_hat*x ##예측값

    err = np.mean((y-y_hat)**2) ##err
    # print(err)

    diff = np.mean(2*x*(y_hat - y))  ##미분값 
    a_hat = a_hat - 1e-4*diff
    print(a_hat)
    y_hat = a_hat*x
    err = np.mean((y-y_hat)**2)
    print(err)
```
